**Spectre de fréquence de classe 00**

```python
import numpy as np
import matplotlib.pyplot as plt

# Charger vos données depuis df
time = df['Time']
acceleration_vertbleu = df['acceleration_vertbleu']
acceleration_vert = df['acceleration_vert']

# Assurez-vous que vos données sont sur un format numérique
time = pd.to_datetime(time)
time_diff = (time - time.iloc[0]).dt.total_seconds()  # Conversion du temps en secondes

# Appliquer la FFT
fft_vertbleu = np.fft.fft(acceleration_vertbleu)
fft_vert = np.fft.fft(acceleration_vert)

# Calculer les fréquences
n = len(time_diff)
frequencies = np.fft.fftfreq(n, d=np.mean(np.diff(time_diff)))  # fréquence d'échantillonnage

# Calculer le spectre (en module)
spectrum_vertbleu = np.abs(fft_vertbleu)
spectrum_vert = np.abs(fft_vert)

# Tracer le spectre superposé
plt.figure(figsize=(10, 6))

# Spectre superposé de 'acceleration_vertbleu' et 'acceleration_vert'
plt.plot(frequencies[:n//2], spectrum_vertbleu[:n//2], label='acceleration_vertbleu', color='blue')
plt.plot(frequencies[:n//2], spectrum_vert[:n//2], label='acceleration_vert', color='red')

# Ajouter des titres et labels
plt.title('Spectres de acceleration_vertbleu et acceleration_vert')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude')

# Ajouter une légende
plt.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
sampling_freq = 25600

# Créer la figure Plotly
fig = go.Figure()

# Boucle sur toutes les colonnes de données
for col in df.columns:
    if df[col].dtype in [np.float64, np.int64]:  # Vérifier que la colonne contient des valeurs numériques
        acceleration_signal = df[col]

        # Appliquer la transformation de Hilbert pour obtenir l'enveloppe
        analytic_signal = hilbert(acceleration_signal)
        envelope = np.abs(analytic_signal)

        # Calcul de la FFT de l'enveloppe
        fft_envelope = np.fft.fft(envelope)
        freq = np.fft.fftfreq(len(envelope), d=1/sampling_freq)

        # Ajouter la trace au graphique
        fig.add_trace(
            go.Scatter(
                x=freq[:len(freq)//2],
                y=np.abs(fft_envelope)[:len(freq)//2],
                mode='lines',
                name=f"FFT - {col}",
                line=dict(width=1.5)
            )
        )

# Ajouter des labels et un titre
fig.update_layout(
    title="Spectre de Fréquence - Enveloppe du Signal",
    xaxis_title="Fréquence (Hz)",
    yaxis_title="Amplitude",
    xaxis=dict(range=[0, 5000]),  # Limiter la plage de fréquence
    template="plotly_white",
    showlegend=True
)

# Afficher le graphique
fig.show()
```

# **3. Calcul des indicateurs statistiques**

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Dossiers contenant les fichiers Excel
folder1 = "/content/drive/MyDrive/Class00_F0/F0"  # Classe 0 : sain
folder2 = "/content/drive/MyDrive/Class02_F0/F0"  # Classe 1 : défaillant

# Fonction d'extraction des features à partir de la colonne 'acceleration_vert'
def extract_features(filepath):
    try:
        df = pd.read_excel(filepath, engine='calamine')
        df = df.drop(columns=["Time"], errors="ignore")  # Supprime la colonne 'Time' si elle existe

        if 'acceleration_vert' not in df.columns:
            print(f"Colonne 'acceleration_vert' manquante dans {filepath}")
            return None

        x = df['acceleration_vert']
        features = {
            'acceleration_vert_mean': x.mean(),
            'acceleration_vert_rms': np.sqrt(np.mean(x**2)),
            'acceleration_vert_skew': skew(x),
            'acceleration_vert_kurtosis': kurtosis(x),
            'acceleration_vert_max': x.max(),
        }

        return features
    except Exception as e:
        print(f"Erreur dans {filepath} : {e}")
        return None

# Fonction pour traiter tous les fichiers d’un dossier donné
def process_folder(folder_path, label):
    feature_list = []
    for file in os.listdir(folder_path):
        if file.endswith(".xlsx") or file.endswith(".xls"):
            path = os.path.join(folder_path, file)
            feats = extract_features(path)
            if feats:
                feats["filename"] = file
                feats["label"] = label
                feature_list.append(feats)
    return pd.DataFrame(feature_list)

# Traitement des deux dossiers
df_class0 = process_folder(folder1, label=0)
df_class1 = process_folder(folder2, label=1)

# Fusion des deux DataFrames
dataset = pd.concat([df_class0, df_class1], ignore_index=True)

# Aperçu du résultat
print(dataset.head())

# Sauvegarde en CSV ou Excel
dataset.to_csv("/content/drive/MyDrive/features_dataset.csv", index=False)
# OU
# dataset.to_excel("/content/drive/MyDrive/features_dataset.xlsx", index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Séparer les features et le label
X = dataset.drop(['label'], axis=1)
y = dataset['label']

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparation en train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

## **5.2 Modèles de Machine Learning (ML)**

**a. Random Forest**
