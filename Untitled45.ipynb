```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Définir les chemins des deux dossiers
folder1 = "/content/drive/MyDrive/Class00_F0/F0"
folder2 = "/content/drive/MyDrive/Class02_F0/F0"

# Fonction pour lire tous les fichiers d'un dossier
def lire_dossier(folder_path):
    data_list = []
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.xlsx'):  # ou .txt selon ton cas
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_excel(file_path , engine='calamine')  # adapter si séparateur est ; ou \t
            data_list.append(df)
    return data_list

# Lecture des deux dossiers
data_class00 = lire_dossier(folder1)
data_class02 = lire_dossier(folder2)
```

```python
# Exemple sur le premier fichier de Class00
df = data_class00[0]
print("Taille du fichier :", df.shape)
print("\nColonnes :", df.columns.tolist())
print("\nTypes de données :")
print(df.dtypes)
print("\nAperçu des données :")
print(df.head())
```

Le premier signal, de faible amplitude (±0.3), présente une forme périodique, régulière et centrée, sans anomalies visibles. Il reflète un comportement mécanique stable, typique d’un roulement sain.

À l’inverse, le deuxième signal montre une amplitude élevée (±2.5), avec des pics impulsionnels et une forte irrégularité. Ces caractéristiques suggèrent la présence d’un défaut mécanique localisé, comme une fissure ou un impact sur un élément roulant.

Ainsi, la comparaison des signaux permet d’identifier les défaillances et de poser les bases d’un système automatisé de maintenance prédictive, renforcé par des analyses fréquentielles et des indicateurs statistiques.

```python
import numpy as np
import matplotlib.pyplot as plt

# Charger vos données depuis df
time = df['Time']
acceleration_vertbleu = df['acceleration_vertbleu']
acceleration_vert = df['acceleration_vert']

# Assurez-vous que vos données sont sur un format numérique
time = pd.to_datetime(time)
time_diff = (time - time.iloc[0]).dt.total_seconds()  # Conversion du temps en secondes

# Appliquer la FFT
fft_vertbleu = np.fft.fft(acceleration_vertbleu)
fft_vert = np.fft.fft(acceleration_vert)

# Calculer les fréquences
n = len(time_diff)
frequencies = np.fft.fftfreq(n, d=np.mean(np.diff(time_diff)))  # fréquence d'échantillonnage

# Calculer le spectre (en module)
spectrum_vertbleu = np.abs(fft_vertbleu)
spectrum_vert = np.abs(fft_vert)

# Tracer le spectre superposé
plt.figure(figsize=(10, 6))

# Spectre superposé de 'acceleration_vertbleu' et 'acceleration_vert'
plt.plot(frequencies[:n//2], spectrum_vertbleu[:n//2], label='acceleration_vertbleu', color='blue')
plt.plot(frequencies[:n//2], spectrum_vert[:n//2], label='acceleration_vert', color='red')

# Ajouter des titres et labels
plt.title('Spectres de acceleration_vertbleu et acceleration_vert')
plt.xlabel('Fréquence (Hz)')
plt.ylabel('Amplitude')

# Ajouter une légende
plt.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```
